```python
dataset_path = Path("csv/spotify_decades_dataset.csv")
dataset_df = pd.read_csv(dataset_path)
dataset_df.head()
```

```python
dataset_df.columns
```

```python
#Drop the additional index columns that were created as part of the appending
dataset_df = dataset_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'])
dataset_df.head()
```

```python
dataset_df.describe()
```

```python
dataset_df['Songs per Year'] = dataset_df.groupby('Year')['Trackname'].transform('count')
dataset_df.tail()
```

```python
#Transform the length of the songs from milliseconds into minutes 
dataset_df['Length']=(dataset_df['Length']/60000).round(4)

#Check out the most popular artists listed in the dataset
dataset_df.value_counts(subset='Artist',ascending=False).head(10)
```

```python
#Acousticness: This value describes how acoustic a song is. A score of 1.0 means the song is most likely to be an acoustic one.
filtered_acoustic = dataset_df.loc[dataset_df['Acousticness']>0.5]
```

```python
#Calculating the mean features over all years
df_mean = dataset_df.groupby('Year').mean().add_suffix(' mean')
df_mean = df_mean.reset_index().astype({'Year':'int'})
df_mean.tail()
```

```python
df_var = dataset_df.groupby('Year').std().add_suffix(' standard deviation')
df_var.reset_index().astype({'Year':'int'})
df_var.head(10)
```

```python
year_features = df_mean.join(df_var, on='Year')
```

```python
year_features.head(10)
```

```python
year_features.to_csv('year_features.csv', index=False)  
```

```python
#Liveness: This value describes the probability that the song was recorded with a live audience. According to the official documentation “a value above 0.8 provides strong likelihood that the track is live”.
filtered_live = dataset_df.loc[dataset_df['Liveness']>0.19]
filtered_live.value_counts(subset='Year').head()
```

```python
#Danceability: “Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable”.
filtered_dance = dataset_df.loc[dataset_df['Danceability']>0.64]
filtered_dance.value_counts(subset='Decade')
```

```python
#Energy: “(energy) represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. Perceptual features contributing to this attribute include dynamic range, perceived loudness, timbre, onset rate, and general entropy.”.
filtered_energy = dataset_df.loc[dataset_df['Energy']>0.80]
filtered_energy.value_counts(subset='Decade')
```

```python
filtered_tempo = dataset_df.loc[dataset_df['Tempo']>120]
filtered_tempo.value_counts(subset='Decade')
```

```python
#Speechiness: “Speechiness detects the presence of spoken words in a track”. If the speechiness of a song is above 0.66, it is probably made of spoken words, a score between 0.33 and 0.66 is a song that may contain both music and words, and a score below 0.33 means the song does not have any speech.
filtered_speech = dataset_df.loc[dataset_df['Speechiness']>0.072]
filtered_speech.value_counts(subset='Decade')
```

```python
#Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content. Values above 0.5 are intended to represent instrumental tracks, but confidence is higher as the value approaches 1.0.
filtered_instrument = dataset_df.loc[dataset_df['Instrumentalness']>0.034]
filtered_instrument.value_counts(subset='Decade')
```

```python
#An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of "3/4", to "7/4".
filtered_time_sig = dataset_df.loc[dataset_df['Time Signature']>4]
filtered_time_sig.value_counts(subset='Decade')
```

```python
#An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of "3/4", to "7/4".
filtered_pop = dataset_df.loc[dataset_df['Popularity']>70]
filtered_pop.value_counts(subset='Decade')
```

```python
#How happy a track is
filtered_valence = dataset_df.loc[dataset_df['Valence']>0.7]
filtered_valence.value_counts(subset='Decade')
```
